```python
import pandas as pd
df = pd.read_csv('jp06.csv')
```

```python
tt=[]
with open('four.txt',encoding="UTF-8") as f:
    for i in f:
      tt.append(i.strip())
```

```python
A=set()
for partcontent in df['content']:
    words=jieba.cut(partcontent)
    for word in words:
        for ltt in tt:
            if word == ltt:
                A.add(ltt)
    if len(A)>1:
        print(A)
    A=set()
```
